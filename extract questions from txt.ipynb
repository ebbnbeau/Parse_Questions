## splitting the file into a small portion just to test the waters

```python
f = open('questions.txt', 'r')

txt = f.read()
txt_split = txt.split('Question:\n')[1]
print(txt_split.split('Hints:\n')[0])
```

### Okay! Let's parse the file into manageable chunks by splitting the text by delimiter of 'Question:'
### Our first split returned us a list, so let's use a for loop to split each string in the list
### Let's also take out hint and solution. We can parse by 'Hints:'
### And create a new txt file to write the results to

```python
f = open('questions.txt', 'r')
f2 = open('result.txt', 'w')

txt = f.read()


def parsing(text):
    splitted = text.split('Question:\n')
    for i in splitted:
        result = i.split('Hints:')[0]
        f2.write(result)

parsing(txt)
```

### Let's take a look at our result

```python
with open('result.txt', 'r') as f:
    for line in f:
        print(line, end = '')
```
